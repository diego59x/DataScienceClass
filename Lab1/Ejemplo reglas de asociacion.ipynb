# Reglas de asociaci√≥n

```python
import pandas as pd
from apyori import apriori   # pip3 install apyori
```

## Cargar los datos

```python
mtcars = pd.read_csv('./data/mtcars.csv', encoding = "ISO-8859-1")
mtcars.head()
```

```python
categoricas = mtcars[['model',"cyl","am","vs","gear"]]
categoricas = categoricas.astype(object)
categoricas
```

## Transformando los datos

```python
print(categoricas.shape)
# Se van a convertir los datos a una lista de listas
records = []
for i in range(0, 32):
    records.append([str(categoricas.values[i,j]) for j in range(0, 5)])
```

```python
records
```

```python
# El m√≠nimo de cobertura o soporte es 20% y el m√≠nimo de confianza es 70%
reglas_asociacion = apriori(records,min_support=0.1, min_confidence = 0.7)
reglas = list(reglas_asociacion)
```

```python
print(len(reglas))
```

Se crearon 16 reglas con los umbrales proporcionados

```python
print(reglas[0])
```

```python
for regla in reglas:
    p = regla[0]
    for e in p:
        print(e)
    #print("La regla: "+elementos)
```
